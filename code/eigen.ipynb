```python
import numpy as np
import matplotlib.pyplot as plt

def plot_vector(vector, color):
    plt.quiver(0, 0, vector[0], vector[1], color=color, angles='xy', scale_units='xy', scale=1)

# Create a 2x2 matrix
A = np.array([[3, 1],
              [1, 2]])

print("Matrix A:")
print(A)

# Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

print("\nEigenvalues:")
print(eigenvalues)

print("\nEigenvectors:")
print(eigenvectors)

# Verify Av = λv for each eigenpair
for i in range(len(eigenvalues)):
    print(f"\nVerifying Av = λv for eigenpair {i + 1}:")
    print(f"A * v{i + 1} =")
    print(np.dot(A, eigenvectors[:, i]))
    print(f"λ{i + 1} * v{i + 1} =")
    print(eigenvalues[i] * eigenvectors[:, i])

# Plotting
plt.figure(figsize=(10, 10))
plt.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
plt.axvline(x=0, color='k', linestyle='-', linewidth=0.5)

# Plot original basis vectors
plot_vector([1, 0], 'gray')
plot_vector([0, 1], 'gray')

# Plot eigenvectors
for i in range(len(eigenvalues)):
    ev = eigenvectors[:, i]
    plot_vector(ev, ['r', 'b'][i])
    plt.text(ev[0], ev[1], f'v{i + 1}', fontsize=12)

# Plot transformed eigenvectors
for i in range(len(eigenvalues)):
    ev = eigenvectors[:, i]
    transformed_ev = np.dot(A, ev)
    plot_vector(transformed_ev, ['darkred', 'darkblue'][i])
    plt.text(transformed_ev[0], transformed_ev[1], f'Av{i + 1}', fontsize=12)

plt.title("Eigenvectors and their transformations")
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)
plt.show()
```
