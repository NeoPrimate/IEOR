```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the variable and the function
x = sp.symbols('x')
f = x**4 - 2*x**2 + 1

# Compute derivatives
f_prime = sp.diff(f, x)
f_double_prime = sp.diff(f_prime, x)
f_triple_prime = sp.diff(f_double_prime, x)
f_quad_prime = sp.diff(f_triple_prime, x)

# List of functions and their derivatives
functions = [f, f_prime, f_double_prime, f_triple_prime]

# Create plots
fig, axs = plt.subplots(len(functions), 1, figsize=(8, 12))

# Define a range for x values
x_vals = np.linspace(-1.5, 1.5, 400)

# Plot each function and its derivative
for i, func in enumerate(functions):
    # Convert sympy function to a numpy function
    func_np = sp.lambdify(x, func, 'numpy')
    y_vals = func_np(x_vals)
    
    # Plot the function
    axs[i].plot(x_vals, y_vals, label=f'${sp.latex(func)}$')
    
    # Compute critical points (roots of the derivative)
    crit_points = sp.solve(func, x)
    crit_points = [float(p) for p in crit_points]
    
    # Plot critical points
    crit_y_vals = [func_np(p) for p in crit_points]
    axs[i].plot(crit_points, crit_y_vals, 'ro')
    
    axs[i].legend()
    axs[i].grid(True)
    axs[i].axhline(0, color='red', linestyle='--')

# Adjust layout and show plot
plt.tight_layout()

plt.savefig('../content/vis/derivatives.png', dpi=300)

plt.show()
```

# Saddle (Critical) Point

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function and its saddle point
def f(x, y):
    return x**2 - y**2

# Create a grid of x and y values
x = np.linspace(-2, 2, 400)
y = np.linspace(-2, 2, 400)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Define saddle point
saddle_x, saddle_y = 0, 0
saddle_z = f(saddle_x, saddle_y)

# Plotting the 3D surface plot
fig = plt.figure(figsize=(14, 6))

# 3D Surface plot
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax1.scatter(saddle_x, saddle_y, saddle_z+1, color='red', s=100, label='Saddle Point')
ax1.set_title('3D Surface Plot of $f(x, y) = x^2 - y^2$')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('f(x, y)')
ax1.legend()

# Contour plot
ax2 = fig.add_subplot(122)
contour = ax2.contourf(X, Y, Z, cmap='viridis', levels=50)
ax2.contour(X, Y, Z, levels=10, colors='k', linewidths=0.5)
ax2.plot(saddle_x, saddle_y, 'ro', markersize=8, label='Saddle Point')
ax2.set_title('Contour Plot of $f(x, y) = x^2 - y^2$')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
fig.colorbar(contour, ax=ax2, orientation='vertical')
ax2.legend()

# Show the plots
plt.tight_layout()

plt.savefig('../content/vis/critical_point_saddle.png', dpi=300)

plt.show()
```

# Fundemantal Theorem of Calculus

```python
import sympy as sp

# Define the variable
x = sp.Symbol('x')

# Define the function f(x)
f = 3 * x**2

# 1. Compute the definite integral of f(x) from 1 to 2

# Find the antiderivative of f(x)
F = sp.integrate(f, x)

# Compute the definite integral using the antiderivative
a, b = 1, 2
definite_integral = F.subs(x, b) - F.subs(x, a)

# Print the result
print("Definite Integral of 3x^2 from 1 to 2:", definite_integral)

# 2. Define the integral function F(x) with limits from 1 to x
F_x = sp.integrate(f, (x, 1, x))

# Compute the derivative of F(x)
F_prime = sp.diff(F_x, x)

# Print the results
print("Integral Function F(x):", F_x)
print("Derivative of the Integral Function F'(x):", F_prime)
```

# Fundemental Theorem of Calculus

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its antiderivative
def f(x):
    return x

def F(x):
    return 0.5 * x**2

# Define the range for x
x = np.linspace(0, 2, 400)
y = f(x)
Y = F(x)

# Plot the function f(x)
plt.figure(figsize=(12, 6))

# Plot f(x) and the shaded area under the curve
plt.subplot(1, 2, 1)
plt.plot(x, y, label='$f(x) = x$')
plt.fill_between(x, y, alpha=0.2)
plt.title('Function $f(x)$ and its Integral')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.legend()
plt.grid(True)

# Plot F(x)
plt.subplot(1, 2, 2)
plt.plot(x, Y, label='$F(x) = \\frac{x^2}{2}$', color='r')
plt.title('Antiderivative $F(x)$')
plt.xlabel('$x$')
plt.ylabel('$F(x)$')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
